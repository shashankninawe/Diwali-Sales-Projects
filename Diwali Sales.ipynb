### data cleaning 

```python
data.drop(['Status','unnamed1'], axis = 1 ,inplace = True)
```

```python
data.head()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

#### We can check that the number of rows has been decreased from 11251 to 11243. This shows tha the duplicate values have been terminated.

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
# Change the data type of Amount and Marital_Status
data['Amount'] = data['Amount'].astype('int64')
data['Amount']
```

```python
# Convert the column into a dataframe so that we could change the values 0,1 with Unmarried and married
data= pd.DataFrame(data)
```

```python
#Change the Marital status values from 0, 1 to unmarried and married.
data['Marital_Status'] = data['Marital_Status'].replace({0: "Unmarried", 1: "Married"})
```

```python
data.head()
```

```python
# change the gender F and M to Male and female
data['Gender'] = data['Gender'].replace({"F":"Female" , "M":"Male"})
```

```python
data.head()
```

```python
data[['Amount','Age','Orders']].describe()
```

### Exploring the data using visual insights

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Gender', data=data , palette='viridis')
for bars in ax.patches:
    ax.annotate(format(bars.get_height()), 
                   (bars.get_x() + bars.get_width() / 2., bars.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 5), 
                   textcoords = 'offset points')

plt.title('Customer Count by Gender', fontsize=13)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
ax = sns.countplot(x='Age Group', data=data , palette='viridis')
for bars in ax.patches:
    ax.annotate(format(bars.get_height()), 
                (bars.get_x() + bars.get_width() / 2., bars.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')

plt.title('Total Customer count of Age Group', fontsize=14)
plt.xlabel('Age group')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Marital_Status', data=data , palette='viridis')  # Changed 'cpintplot' to 'countplot'
for bars in ax.patches:
    ax.annotate(format(bars.get_height()),
                (bars.get_x() + bars.get_width() / 2., bars.get_height()),  # Fixed typo in 'grt_height' to 'get_height'
                ha="center", va="center",
                xytext=(0, 4),
                textcoords='offset points')

plt.title('Customer count by Marital_status', fontsize=14)
plt.xlabel('Marital_status')
plt.ylabel('count')  
plt.show()
```

```python
plt.figure(figsize=(15, 4))
Total_revenue_age = data.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Age Group', y='Amount' , data=Total_revenue_age , palette='viridis') 
plt.title('Total Revenue Generated by each Age-Group', fontsize=13)
plt.xlabel('Age Group')
plt.ylabel('Amount')  
plt.show()
```

```python
plt.figure(figsize=(12, 5))
ax = sns.countplot(x='Age Group', data=data , hue = 'Gender' , palette='viridis')  # Changed 'cpintplot' to 'countplot'
for bars in ax.patches:
    ax.annotate(format(bars.get_height()),
                (bars.get_x() + bars.get_width() / 2., bars.get_height()),  # Fixed typo in 'grt_height' to 'get_height'
                ha="center", va="center",
                xytext=(0, 4),
                textcoords='offset points')

plt.title('Total Customer Count Vs Age-Group Based on Gender', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('count')  
plt.show()
```

```python
grouped_data = data.groupby('Age Group')
grouped_data
```

```python
sum_of_revenue = grouped_data["Amount"].sum()
sum_of_revenue
```

```python
plt.figure(figsize=(15,7))
plt.pie(sum_of_revenue, labels = grouped_data["Age Group"].unique(), autopct="%1.1f%%")
plt.title("Contribution to Revenue by Age-Group" , fontsize = 14)
plt.show()
```

### State

- We will analyze the data to get the top 5 States that have highest sales

```python
data.head()
```

```python
grouped_state = data.groupby("State").sum("Amount")
print(grouped_state)
```

```python
Total_revenue_State = data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=(15, 4))
ax = sns.barplot(x='State', y='Amount' , data=Total_revenue_State , palette='viridis') 
plt.title('Total Revenue Generated by Each State', fontsize=13)
plt.xlabel('State',fontsize=14)
plt.ylabel("Amount", fontsize=14)  
plt.xticks(rotation = 90)
plt.show()
```

```python
Total_orders_state = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
plt.figure(figsize=(12, 8))
states_plot = sns.barplot(x = 'Orders', y = 'State', data = Total_orders_state , palette='viridis')
plt.title("Total orders by each State", fontsize=16)
plt.ylabel("States", fontsize=14)

for bars in states_plot.containers:
    states_plot.bar_label(bars)
```

### Product Category¶
In order to see the sales of each category product we will use the data and check for the top 5 categories of product which has highest sales and which are contributing most to the total sales.

- Plot a column graph to see which products have highest orders

```python
data.head()
```

```python
# plot a column graph to see which products have highest orders

plt.figure(figsize=(15, 4))
sns.countplot(x = 'Product_Category', data = data , palette='viridis')
plt.xlabel("Product Category")
plt.ylabel("Number of orders")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(x='Product_Category', data=data , hue = 'Gender' , palette='viridis')  # Changed 'cpintplot' to 'countplot'
plt.title('Total order count for each product on basis of Gender', fontsize=14)
plt.xlabel('Product_Category')
plt.ylabel('count')  
plt.xticks(rotation = 90)
plt.show()
```

##### Now will graph the data to see which products are generating most sales revenue

```python
Product = data.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
```

```python
plt.figure(figsize=(15, 4))
ax = sns.barplot(x='Product_Category', y='Amount' , data = Product , palette='viridis') 
for bars in ax.patches:
    ax.annotate(format(bars.get_height()),
                (bars.get_x() + bars.get_width() / 2., bars.get_height()),  # Fixed typo in 'grt_height' to 'get_height'
                ha="center", va="center",
                xytext=(0, 5),
                textcoords='offset points')
    
plt.title('TOTAL REVENUE GENERATED BY TOP 5 PRODUCTS', fontsize=13)
plt.xlabel('Product_Category')
plt.ylabel('Amount')  
plt.show()
```

## Occupation¶
##### Now we will see persons from which occupation is more likely to praticipate in diwali sales offer.

- Graph a column chart for Total Cutomer count by customers profession

```python
data.head()
```

```python
Occupation = data.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=(12, 5))
Occupation_plot = sns.countplot(x='Occupation', data=data , palette='viridis')   # Changed 'cpintplot' to 'countplot'
plt.title('Total Customer count by Occupation', fontsize=14)
plt.xlabel('Occupation')
plt.ylabel('count')  
plt.xticks(rotation = 90)
plt.show()

for bars in Occupation_plot.containers:
    Occupation_plot.bar_label(bars)
```

#### Now we will check the top 5 Occupations that are contributing most to the revenue

```python
Occupation_df = data.groupby(['Occupation', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
```

```python
plt.figure(figsize=(15, 4))
ax = sns.barplot(x='Occupation', y='Amount' , data = Occupation_df , palette='viridis') 
for bars in ax.patches:
    ax.annotate(format(bars.get_height()),
                (bars.get_x() + bars.get_width() / 2., bars.get_height()),  # Fixed typo in 'grt_height' to 'get_height'
                ha="center", va="center",
                xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Top 5 Revenue Generated  Vs Occupation of customers', fontsize=13)
plt.xlabel('Occupation')
plt.ylabel('Amount')  
plt.show()
```

```python
plt.figure(figsize=(15, 4))
ax = sns.barplot(x='Occupation', y='Amount' , hue = 'Gender', data = Occupation_df , palette='viridis') 
for bars in ax.patches:
    ax.annotate(format(bars.get_height()),
                (bars.get_x() + bars.get_width() / 2., bars.get_height()),  # Fixed typo in 'grt_height' to 'get_height'
                ha="center", va="center",
                xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Top 5 Revenue Generated  Vs Occupation of customers', fontsize=13)
plt.xlabel('Occupation')
plt.ylabel('Amount')  
plt.show()
```
